# Multiple Linear Regression

Now we learnt how to build a model using one X(feature variable) and Y(response variable) but what happens if we have multiple feature variables? 

It is not possible to make a separate model for each,combining them and then understanding them.Thus we use Multiple Linear Regression

## Understanding the data

Looks like TV and Radio has some correlation with Sales but Newspaper and Sales dont have much correlation

**1. Splitting the Data for Training and Testing**

```python
#putting all feature variables in x
x=df[['TV','Radio','Newspaper']]

#putting the label (response variable) in y
y=df['Sales']
```

```python
from sklearn.cross_validation import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=100)
```

**2. Performing Linear Regression**


```python
#import library
from sklearn.linear_model import LinearRegression

#create object
lr=LinearRegression()

#fit the model to training data
lr.fit(x_train,y_train)
```

**3. Model Evaluation**

```python
print(lr.intercept_)
```

```python
coeff=pd.DataFrame(lr.coef_,x_test.columns,columns=['Coeff'])
coeff
```

Thus if we increase value of TV by 1 unit Sales will be affected by 0.045 units

y=0.045TV+0.189Radio+0.004Newspaper+2.65 is our equation now which we will use to predict our Sales

**4. Predictions**

```python
y_pred=lr.predict(x_test)
y_pred
```

**5. Error Terms**

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse=mean_squared_error(y_test,y_pred)
r_squared=r2_score(y_test,y_pred)

print 'Mean Squared Error:',mse
print 'R Square Value:',r_squared
```

Mse is 1.85 means the model is not able to match only 1% of the values
R2 value between 60-70 is a decent model. This is a very good model.

**6. P value**

```python
import statsmodels.api as sm
```

```python
x_train_sm=x_train
#Unline sklearn,statsmodel dont automatically fit a constant,so we add it using sm.add_constant(x)
x_train_sm=sm.add_constant(x_train_sm)

#fit it to the model
lm_l=sm.OLS(y_train,x_train_sm).fit()

#print coeff
lm_l.params
```

These are the same values that we got from coefficients

```python
print(lm_l.summary())
```

Interpretations:
1. Higher the coef higher will be the std error
2. Lower the coef lower will be the std error
3. T value is coef divided by std error


HO Null hypothesis= Variable is not going to be significant  
H1 Alternate hypothesis=Variable is going to be significant

if pvalue > 0.05 the we fail to reject the null hypothesis ie. Variable is not going to be significant will be true.

if pvalue < 0.05 then we reject null hypothesis i.e Variable is not going to be significant is false i.e Variable is significant infact.

**Therefore newspaper with pvalue 0.541 indicates that this variable is not significant**

**7. Removing Newspaper and running the model again**

```python
#Removing newspaper
x_train_new=x_train[['TV','Radio']]
x_test_new=x_test[['TV','Radio']]
```

```python
#Model
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x_train_new,y_train)
```

```python
#Predicting
y_pred_new=lm.predict(x_test_new)
y_pred
```

```python
print(x_test_new.shape)
```

```python
#Actual VS Predicted
import matplotlib.pyplot as plt
c=[i for i in range(1,61,1)] #creating index as x_test had 60 values
fig=plt.figure()
plt.plot(c,y_test,color="blue",linewidth=2.5,linestyle="-")
plt.plot(c,y_pred,color="red",linewidth=2.5,linestyle="-")
fig.suptitle('Actual vs Predicted',fontsize=20)
plt.xlabel('Index',fontsize=18)
plt.ylabel('Sales',fontsize=18)
```

Here red represents the prediction value and blue represents the actual value. As these are pretty close it means that the model is good.

```python
#Error terms
c=[i for i in range(1,61,1)]
error=y_test-y_pred
fig=plt.figure()
plt.plot(c,error,color='blue',linewidth=3,linestyle='-')
fig.suptitle('Error Terms',fontsize=20)
plt.xlabel('Index',fontsize=18)
plt.ylabel('Error=(y_test)-(y_pred)',fontsize=18)
```

Sketch a line from 0 on the y axis ie. Error and compare the peaks. Only some values are very far from this line i.e have maximum error means the predictions are not close for these values.

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse=mean_squared_error(y_test,y_pred_new)
r_squared=r2_score(y_test,y_pred_new)

print 'Mean Squared Error:',mse

print 'R Square Value:',r_squared
```

Hence the model is even better now with R square value of 0.90
